```python
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt 
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv(Path('./Resources/Salary_Data.csv'))
df.head()
```

```python
#create a visual relationship between years of experience and salary
plt.scatter(df.YearsExperience, df.Salary)
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

```python
#reshape data for scikit-learn. First arg specifies number of rows. -1 means unspecified. numpy will automatically identify number of rows. second arg refers to number of columns or independent variables
X = df.YearsExperience.values.reshape(-1,1)
#output is a 2D array
X[:5]
```

```python
#the shape of X is 30 samples with a single feature (column)
X.shape
```

```python
#assign target variable
y = df.Salary
```

```python
#create instance of linear regression model
model = LinearRegression()
```

```python
#learning stage (fitting or training)
#by convention X is captialized and y is lowercase
model.fit(X,y)
```

```python
#generate predictions
y_pred = model.predict(X)
print(y_pred.shape)
```

```python
#plot predictions against data points
plt.scatter(X,y)
plt.plot(X, y_pred, color='red')
plt.show()
```

```python
#print slope and y-intercept
print(model.coef_)
print(model.intercept_)
```
